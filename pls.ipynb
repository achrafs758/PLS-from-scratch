```python
corr = df.corr()
corr
```

```python
# normalize data
from sklearn import preprocessing
df_scaled = preprocessing.scale(df)
```

```python
df_scaled
```

```python
corr_y = corr.iloc[7][0:7]
corr_yabs = abs(corr_y)
```

```python
corr_yabs
```

```python
Xi = []
for i, elem in enumerate(corr_yabs):
    if elem > 0.5:
        Xi.append(i)
```

```python
Xi
```

```python
df_scaled = pd.DataFrame(df_scaled)
```

```python
import math
sum1 = 0
sum2 = 0
for i in Xi:
    sum1 += corr["Y"][:-1][i] * df_scaled[i]
    sum2 += corr["Y"][:-1][i]**2
```

```python
T1 = (1 / math.sqrt(sum2)) * sum1
```

```python
import statsmodels.api as sm

Y_scaled = df_scaled[7]
```

```python
model = []
for i in range(7):
    X_ = pd.concat([df_scaled[i], T1], axis=1)
    model.append(sm.OLS(Y_scaled, X_).fit())
```

```python
col2 = []
for i in range(len(df_scaled.columns)-1):
    if i == 0:
        if model[i].pvalues[i:i+1][0] < 0.05:
            col2.append(i)
    else:
        if model[i].pvalues[i] < 0.05:
            col2.append(i)
```

```python
col2
```

```python
import numpy as np

residu=[]    
for i in col2:
    dt=pd.DataFrame(T1)
    model = sm.OLS(df_scaled[i], dt).fit()
    residu.append(model.resid) 
X1n=[]
for r in residu:
    X1n.append(r/np.var(r))
```

```python
model2 = []
for i, l in enumerate(X1n):
    dt=pd.DataFrame(T1)
    dt[1]=X1n[i]
    model2.append(sm.OLS(Y_scaled, dt).fit())
```

```python
coefs = np.array([model2[i].params[1:2] for i in range(len(X1n))])
```

```python
coefs
```

```python
sum1 = 0
sum2 = 0
for i in range(len(residu)):
    sum1 += coefs[i] * residu[i]
    sum2 += coefs[i]**2
```

```python
T2 = sum1 / (math.sqrt(sum2))
```
